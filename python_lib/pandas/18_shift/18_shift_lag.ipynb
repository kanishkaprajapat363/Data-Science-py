```python
import pandas as pd
df = pd.read_csv("fb.csv",parse_dates=['Date'],index_col='Date')
df
```

<h2 style="color:purple">Shift</h2>

```python
df.shift(1)
#Move Data Up/Down
```

```python
df.shift(-1)
```

```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
df['Price Change'] = df['Price'] - df['Prev Day Price']
df
```

```python
df['5 day return'] =  (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```

```python
df = df[['Price']]
df
```

<h2 style="color:purple">tshift</h2>Move the Time Index, Not the Data

```python
df.index
```

```python
df.index = pd.date_range(start='2017-08-15',periods=10, freq='B')
df
```

```python
df.index
```

```python
df.tshift(1)
```

| Operation     | Affects       | Common Use                   |
| ------------- | ------------- | ---------------------------- |
| `shift(n)`    | Data values   | Lag features, percent change |
| `shift(freq)` | Index (dates) | Time-based forecasting shift |
